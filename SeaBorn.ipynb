```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load dataset

tips_val = sns.load_dataset('tips')
tips_val
```

```python
tips_val.info()
```

```python
tips_val.describe()
```

```python

sns.set_style('darkgrid')
sns.violinplot(x='total_bill',data = tips_val)
plt.title('TOTAL_BILL')
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris['species']
```

```python
sns.swarmplot(x='species',y= 'petal_length',data=iris)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
titanic.info()
```

```python
#titanic.dropna() - if na value is present. In this case there no need to do drop
```

```python
t = sns.factorplot("class",'survived','sex',data=titanic,kind = 'bar',palette = 'muted',legend = True)
plt.show()
#legend shows description / notation of graph that is green represents female and blue represents male,
#that is shown in right side box
```

```python
t = sns.factorplot("day",'total_bill','smoker',data=tips_val,kind = 'bar',palette = 'muted',legend = True)
plt.show()
```

```python
sns.relplot(x="total_bill",y='tip',hue='smoker',style='time',data=tips_val)
plt.show()
```

```python
sns.relplot(x="sepal_length",y='sepal_width',hue='species',style='species',data=iris)
plt.show()
```

```python
tips = sns.load_dataset("tips")
```

```python
bx = sns.boxplot(x='total_bill',data=tips)
bx.set_title("Boxplot")
plt.show()
```

```python
import numpy as np 
import pandas as pd
x = 10 ** np.arange(1,10)
y = x * 2

data =pd.DataFrame(data={'x':x,'y':y})

# lmplot

lmgrid = sns.lmplot('x','y',data,size = 7,scatter_kws = {'s':100})
lmgrid.set_xticklabels(rotation = 30)
lmgrid.set_yticklabels(rotation = 30) #change the degree of x and y labels

plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',hue='size',size='size',data=tips)
plt.show()
```
